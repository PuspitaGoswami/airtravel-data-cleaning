```python
df.columns = df.columns.str.replace('"','').str.strip()
df.columns
```

```python
df["Month"].head()
```

```python
df["Month"] = pd.to_datetime(df["Month"],format="%b").dt.month_name()
df["Month"]
```

```python
df_long =  df.melt(id_vars="Month", var_name="Year", value_name="Passengers")
df_long
```

```python
df_long["Month"] = pd.Categorical(df_long["Month"],  ["January","February","March","April","May","June",
     "July","August","September","October","November","December"],
      ordered=True)
df_long = df_long.sort_values(["Month","Year"])
df_long
```

```python
df_long.to_csv("cleaned_airtravel.csv",index=False)
```
